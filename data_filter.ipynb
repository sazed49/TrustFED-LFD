```python
import os
import shutil

SRC_DIR = "S:/Summer25/MineDataset/Annotation_Done/MAIN/Frame_4_YOLO/obj_train_data/val"

WITH_DIR = "S:/Summer25/MineDataset/Annotation_Done/With_Lables/Frame_2_with_Labels/with_labels/val"
EMPTY_DIR = "S:/Summer25/MineDataset/Annotation_Done/With_Lables/Frame_2_with_Labels/empty/val"

os.makedirs(WITH_DIR, exist_ok=True)
os.makedirs(EMPTY_DIR, exist_ok=True)

files = os.listdir(SRC_DIR)

images = [f for f in files if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

for img in images:
    name = os.path.splitext(img)[0]
    label = name + ".txt"

    img_path = os.path.join(SRC_DIR, img)
    label_path = os.path.join(SRC_DIR, label)

    # Check annotation
    if os.path.exists(label_path) and os.path.getsize(label_path) > 0:
        shutil.copy(img_path, os.path.join(WITH_DIR, img))
        shutil.copy(label_path, os.path.join(WITH_DIR, label))
    else:
        shutil.copy(img_path, os.path.join(EMPTY_DIR, img))
        if os.path.exists(label_path):
            shutil.copy(label_path, os.path.join(EMPTY_DIR, label))

print("✅ Separation complete")
```
