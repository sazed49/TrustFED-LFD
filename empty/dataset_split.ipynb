```python
import os
import shutil
import random

SRC_DIR = "S:/Summer25/MineDataset/Annotation_Done/MAIN/Frame_5_YOLO/obj_train_data"
OUT_TRAIN = "S:/Summer25/MineDataset/Annotation_Done/MAIN/Frame_5_YOLO/obj_train_data/train"
OUT_VAL = "S:/Summer25/MineDataset/Annotation_Done/MAIN/Frame_5_YOLO/obj_train_data/val"

TRAIN_RATIO = 0.8
RANDOM_SEED = 42

os.makedirs(OUT_TRAIN, exist_ok=True)
os.makedirs(OUT_VAL, exist_ok=True)

random.seed(RANDOM_SEED)

pairs = []

# Collect valid image–label pairs
for f in os.listdir(SRC_DIR):
    if f.lower().endswith(('.jpg', '.png', '.jpeg')):
        name = os.path.splitext(f)[0]
        label_path = os.path.join(SRC_DIR, name + ".txt")

        if os.path.exists(label_path):
            with open(label_path, "r") as lf:
                if any(line.strip() for line in lf):
                    pairs.append(name)

random.shuffle(pairs)
split_idx = int(len(pairs) * TRAIN_RATIO)

train_names = pairs[:split_idx]
val_names = pairs[split_idx:]

def move_pair(name, dst):
    for ext in ['.jpg', '.png', '.jpeg', '.txt']:
        src = os.path.join(SRC_DIR, name + ext)
        if os.path.exists(src):
            shutil.move(src, os.path.join(dst, name + ext))

for name in train_names:
    move_pair(name, OUT_TRAIN)

for name in val_names:
    move_pair(name, OUT_VAL)

print(f"✅ Moved {len(train_names)} train / {len(val_names)} val samples")
```
