```python
import os
from pathlib import Path

VIDEOS_DIR = "S:/Summer25/MineDataset/Annotation_Done/With_Lables"  # parent folder containing video1, video2, ...

video_folders = sorted([f for f in Path(VIDEOS_DIR).iterdir() if f.is_dir()])

for vid_idx, video_folder in enumerate(video_folders, start=1):
    prefix = f"video{vid_idx}"

    # Loop over both train and val subfolders
    for subfolder in ["train", "val"]:
        folder_path = video_folder / subfolder
        if not folder_path.exists():
            print(f"[WARNING] {folder_path} does not exist, skipping.")
            continue

        images = sorted(list(folder_path.glob("*.jpg")))
        labels = sorted(list(folder_path.glob("*.txt")))
        assert len(images) == len(labels), f"Mismatch in {folder_path}"

        for idx, (img, lbl) in enumerate(zip(images, labels), start=1):
            new_name = f"{prefix}_{idx:05d}"  # e.g., video1_00001
            new_img = folder_path / f"{new_name}.jpg"
            new_lbl = folder_path / f"{new_name}.txt"
            os.rename(img, new_img)
            os.rename(lbl, new_lbl)

        print(f"[INFO] Renamed {len(images)} files in {folder_path}")
```
